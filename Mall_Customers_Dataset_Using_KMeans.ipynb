```python
x=df.iloc[:,3:].values
x
```

```python
from sklearn.cluster import KMeans
```

```python
cs=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
  kmeans.fit(x)
  cs.append(kmeans.inertia_)
plt.plot(cs)
```

```python
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=0)
y_means=kmeans.fit_predict(x)
```

```python
kmeans.inertia_
```

```python
labels=kmeans.labels_
print(labels)
```

```python
y_means
```

```python
plt.scatter(x[y_means==0,0],x[y_means==0,1],color='red',label='cluster1')
plt.scatter(x[y_means==1,0],x[y_means==1,1],color='yellow',label='cluster2')
plt.scatter(x[y_means==2,0],x[y_means==2,1],color='orange',label='cluster3')
plt.scatter(x[y_means==3,0],x[y_means==3,1],color='blue',label='cluster4')
plt.scatter(x[y_means==4,0],x[y_means==4,1],color='green',label='cluster5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='red',label='centroids')
```
